```python
import pandas as pd

file_path = 'datos_limpios.csv'
data = pd.read_csv(file_path)

data.head(), data.info()
```

```python
import matplotlib.pyplot as plt

emisiones_por_año = data.loc[:, "1970":"2023"].sum()

plt.figure(figsize=(12, 6))
plt.plot(emisiones_por_año.index, emisiones_por_año.values, marker='o', color='b', label='Total emisiones')
plt.title('Evolución de las Emisiones Totales (1970-2023)', fontsize=14)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Emisiones Totales', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

paises_interes = [
    "Aruba", "Anguilla", "Antigua and Barbuda", "American Samoa", "Argentina",
    "Bahamas", "Barbados", "Belize", "Bermuda", "Bolivia", "Brazil", "Canada",
    "Chile", "Colombia", "Cuba", "Cayman Islands", "Dominican Republic",
    "Ecuador", "French Guiana", "Guatemala", "Guyana", "Honduras", "Jamaica",
    "Mexico", "Nicaragua", "Panama", "Peru", "Puerto Rico", "Suriname",
    "Trinidad and Tobago", "United States", "Venezuela", "Virgin Islands (British)",
    "Saint Vincent and the Grenadines"
]

filtered_data = data[data['pais_nombre'].isin(paises_interes)]

annual_emissions = filtered_data.groupby('pais_nombre').sum(numeric_only=True).T

plt.figure(figsize=(12, 8))

for country in annual_emissions.columns:
    plt.plot(annual_emissions.index, annual_emissions[country], label=country, alpha=0.7)

plt.title("Evolución de las Emisiones por País en América (1970-2023)", fontsize=14)
plt.xlabel("Año", fontsize=12)
plt.ylabel("Emisiones Totales", fontsize=12)
plt.legend(loc='upper left', fontsize=8, ncol=2, frameon=False)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'datos_limpios.csv'
data = pd.read_csv(file_path)

paises_asia = [
     "Afghanistan", "Armenia", "Azerbaijan", "Bahrain", "Bangladesh", "Bhutan",
    "Brunei Darussalam", "Cambodia", "China", "Georgia", "India", "Indonesia",
    "Iran", "Iraq", "Israel", "Japan", "Jordan", "Kazakhstan", "Kuwait",
    "Kyrgyzstan", "Lebanon", "Malaysia", "Maldives", "Mongolia", "Myanmar",
    "Nepal", "Oman", "Pakistan", "Palau", "Philippines", "Qatar", "Saudi Arabia",
    "Singapore", "Sri Lanka", "Syria", "Tajikistan", "Taiwan", "Thailand",
    "Timor-Leste", "Turkmenistan", "United Arab Emirates", "Uzbekistan",
    "Vietnam", "Yemen"
]

asia_data = data[data["pais_nombre"].isin(paises_asia)]

emisiones_actividades = asia_data.groupby("origen_emision").sum().loc[:, "1970":"2023"]

emisiones_actividades.T.plot(kind="bar", stacked=True, figsize=(14, 7), colormap="viridis")
plt.title('Evolución de las Emisiones por Actividad en Asia (1970-2023)', fontsize=14)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Emisiones Totales', fontsize=12)
plt.legend(title='Actividad', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
